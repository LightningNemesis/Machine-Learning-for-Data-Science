```python
import pandas as pd

# Reading data from csv file into a data frame without specifying header
df = pd.read_csv("salaries.csv", header=None)

# Setting 1st row as header (column names of df)
df.columns = df.iloc[0]

# keys = df.iloc[0].tolist()
# py_dict = {key: None for key in keys}

# print(py_dict)

# Dropping 1st row since we are using it as header
df = df[1:].reset_index(drop=True)

# Setting PlayerID column as index column
df.set_index('playerID', inplace=True)

# converting salary to int
df['salary'] = df['salary'].astype(int)

# print("\nDataFrame after setting the first row as header:")
# print(df)
# print('-' * 40)

# Select ID of players: registerd in ATL or HOU && salary >= 1 million
filtered_ATL_OR_HOU_GREATER_THAN_MILLION_df = df[(df['teamID'].isin(['ATL', 'HOU'])) & (df['salary'] > 1000000)]

# print("\nDataFrame after filtering:")
# print(filtered_ATL_OR_HOU_GREATER_THAN_MILLION_df)
# print('-' * 40)

'''
Use describe to calculate:

a. standard deviation
b. first quartile
c. median
d. third quartile
e. mean
f. maximum
g. minimum 

of the salary in team ATL
'''

''' Manually calling funcs. to get each statistical value'''

# Filtering players with teamID as ATL
filtered_ATL_df = df[df['teamID'] == 'ATL']

# mean of salary
mean = filtered_ATL_df['salary'].mean()

# median of salary (odd data count: middle value, even data count: average of 2 middle values)
median = filtered_ATL_df['salary'].median()

# standard deviation (+ - of this value from mean, represents 68.26% of salary data)
std_dev = filtered_ATL_df['salary'].std()

# first quartile (salary below which 25% of the data exists)
first_quartile = filtered_ATL_df['salary'].quantile(0.25)

# third quartile (salary below which 75% of the data exists)
third_quartile = filtered_ATL_df['salary'].quantile(0.75)

maximum = filtered_ATL_df['salary'].max()
minimum = filtered_ATL_df['salary'].min()

print("mean", mean)
print("standard dev", std_dev)
print("median", median)
print("first quartile", first_quartile)
print("third quartile", third_quartile)
print("maximum", maximum)
print("minimum", minimum)

''' Using Describe to obtain same values '''
print('-'*40)
print(filtered_ATL_df['salary'].describe())
```

```python
''' f.1) Converting df to dict using iterrows() '''

# Reading data from csv file into a data frame without specifying header
df = pd.read_csv("salaries.csv", header=None)

# Setting 1st row as header (column names of df)
df.columns = df.iloc[0]

keys = df.iloc[0].tolist()
py_dict = {key: [] for key in keys}

# Dropping 1st row since we are using it as header
df = df[1:].reset_index(drop=True)

for index, row in df.iterrows():
    py_dict['yearID'].append(row['yearID'])
    py_dict['teamID'].append(row['teamID'])
    py_dict['lgID'].append(row['lgID'])
    py_dict['playerID'].append(row['playerID'])
    py_dict['salary'].append(row['salary'])

# print(py_dict)
```

```python
''' f.2) Converting df to dict using to_dict() '''

# Convert the DataFrame to a dictionary (default orientation)
df_dict = df.to_dict()

# print("converted dict")
# print(df_dict["salary"])
```

```python
''' g) Converting dict to data frame & updating header of data frame '''
df_from_dict = pd.DataFrame(df_dict)
df_from_dict.columns = ['a', 'b', 'c', 'd', 'e']

# print(df_from_dict)
```
